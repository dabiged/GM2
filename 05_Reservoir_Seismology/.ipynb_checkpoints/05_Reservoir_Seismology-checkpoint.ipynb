## Part 2: Shear stimulation of faults


### a) Determine the seismic moment

Determine the seismic moment, $M_0$ , of the events in each well from the moment
magnitudes, $M_w$ , using the relationship below.

$$ M_w=\frac{2}{3} (log_{10}M_0-9)$$

$$ \frac{3M_w}{2}= log_{10}M_0-9$$

$$ \frac{3M_w}{2} +9 = log_{10}M_0$$

$$ 10 ^{\frac{3M_w}{2} +9} = M_0$$

```python
MoMagnitude_A=np.power(10,((3*Magnitude_A/2)+9))
MoMagnitude_B=np.power(10,((3*Magnitude_B/2)+9))
MoMagnitude_C=np.power(10,((3*Magnitude_C/2)+9))
MoMagnitude_D=np.power(10,((3*Magnitude_D/2)+9))
```

```python
print("Total Moment Magnitude in Well B", MoMagnitude_B.cumsum()[len(MoMagnitude_B)-1])
print("Total Moment Magnitude in Well D", MoMagnitude_D.cumsum()[len(MoMagnitude_D)-1])
```

### b) Calculate the total surface area created by the microseismic events associated with stimulation in each well. 

Assume a unifrom stress drop for all events $\Delta \tau = 0.5 MPa$. Use the equation below to calculate the surface area created by each microseismic event, which assumes a circular fault patch with radius, r. Please see Unit 12 Section 3 for details.

$$ \pi r^2 = S = \pi (\frac {7M_0}{16\Delta\tau})^\frac{2}{3} $$


```python
Area_A=np.pi*np.power((7*MoMagnitude_A)/(16*0.5*10.0**6.),2/3)
Area_B=np.pi*np.power((7*MoMagnitude_B)/(16*0.5*10.0**6.),2/3)
Area_C=np.pi*np.power((7*MoMagnitude_C)/(16*0.5*10.0**6.),2/3)
Area_D=np.pi*np.power((7*MoMagnitude_D)/(16*0.5*10.0**6.),2/3)
```
