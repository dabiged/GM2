## Part 1: Creep and stress relaxation
The time-dependent deformation (creep) of unconventional reservoir rocks can be described in
terms of a viscoelastic power law with the form:
$$ \epsilon_{creep}(t) = B \sigma t^n$$

### a) 
How are power law parameters B and n expected to vary with clay + kerogen content and
sample orientation?

B is the elastic compliance which is akin to the inverse of the Young's Modulus, and,
n is the time dependent creep.

Clay+Kerogen Content:
Increase in Clay and Kerogen would see:
an increase in the elastic compliance
an increase in the time dependent creep

Sample Orientation:
Vertical wells (i.e. perpendiclar to strata) would creep more, so a higher value of n.
Vertical wells (i.e. perpendiclar to strata) would have more compliance (higher value of B).

### d.
Re-arrange the Eqn. 1 to obtain the expression below for the creep compliance function
J(t). Plot log J(t) vs. log t for each sample. Show how the values of B and n are obtained
from this plot.
$$log J (t) = log B + nlogt$$

$$log ( \epsilon_{creep}(t)) = log (B \sigma t^n )$$

$$log ( J(t)) = log (B ) + log ( \sigma ) + n*log (t) )$$

The value sigma is set as the unit stress.
$$log ( J(t)) = log (B ) + n*log (t) )$$

Ploting the log of J(t) on the y-axis and the log(t) on the x-axis gives:
n as the gradient, and,
B as the y-intercept.
